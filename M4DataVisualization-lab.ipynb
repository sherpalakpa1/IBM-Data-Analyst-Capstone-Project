## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing distribution of data

Plot a histogram of `ConvertedComp.`

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python


# your code goes here
QUERY = """
SELECT *
FROM DevType
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
dev_df = pd.read_sql_query(QUERY,conn)
dev_df.value_counts()
dev_df = dev_df.groupby('DevType')['Respondent'].count().reset_index().sort_values('Respondent',ascending=False)
dev_df
```

```python

# your code goes here
QUERY = """
SELECT *
FROM PlatformWorkedWith
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
db_df = pd.read_sql_query(QUERY,conn)
db_df = db_df.groupby('PlatformWorkedWith')['Respondent'].count().reset_index().sort_values('Respondent',ascending=False)
db_df
```

```python

# your code goes here
QUERY = """
SELECT *
FROM DatabaseWorkedWith
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
db_df = pd.read_sql_query(QUERY,conn)
db_df = db_df.groupby('DatabaseWorkedWith')['Respondent'].count().reset_index().sort_values('Respondent',ascending=False)
db_df
```

```python

# your code goes here
QUERY = """
SELECT *
FROM LanguageWorkedWith
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
lan_df = pd.read_sql_query(QUERY,conn)
lan_df = lan_df.groupby('LanguageWorkedWith')['Respondent'].count().reset_index().sort_values('Respondent',ascending=False)
lan_df
```

```python
# your code goes here
QUERY = """
SELECT *
FROM LanguageDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
lan_df = pd.read_sql_query(QUERY,conn)
lan_df = lan_df.groupby('LanguageDesireNextYear')['Respondent'].count().reset_index().sort_values('Respondent',ascending=False)
lan_df.head()
```

```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
data_df = pd.read_sql_query(QUERY,conn)
data_df = data_df.groupby('DatabaseDesireNextYear')['Respondent'].count().reset_index().sort_values('Respondent',ascending=False)
data_df = data_df.head()
plt.figure(figsize=(10,10))
plt.pie(data=data_df,x='Respondent',labels='DatabaseDesireNextYear',autopct='%.2f%%')
plt.title('Top 5 Databases that Respondents Wish To Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# your code goes here
df_30to35= df[(df['Age']>=30) & (df['Age']<=35)]
df_30to35 = df_30to35.groupby('Age')[['WorkWeekHrs','CodeRevHrs']].median()

df_30to35.plot(kind='bar',stacked=True,figsize=(10,7))
plt.xlabel('Age')
plt.ylabel('Hrs')
plt.title('The Median WorkWeekHrs and CodeRevHrs for all ages from 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_45to60 = df[(df['Age']>=45) & (df['Age']<=60)]
df_45to60 = df_45to60.groupby('Age')['ConvertedComp'].median().reset_index()
df_45to60.plot(kind='line',x='Age',y='ConvertedComp',figsize=(10,7))
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.title('The Median Converted Comp for all ages from 45 to 60')
plt.show()
```
